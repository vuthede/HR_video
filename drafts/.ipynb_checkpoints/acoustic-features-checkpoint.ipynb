```python
import audiosegment
import pydub 
import numpy as np
import sys
modulePath = '../ChristiansPythonLibrary/src' 
sys.path.append(modulePath)
import generalUtility
import dspUtil
import praatUtil
import generalUtility
import matplotlibUtil

import parselmouth 


sound = audiosegment.from_file("ses01.wav")
samples = sound.get_array_of_samples()
rate = sound.frame_rate
print(rate)
voiced = sound.filter_silence(duration_s=0.2)
samples = voiced.get_array_of_samples()

def amp2Db(samples):
    dbs = []
    for  x in samples:
        if x < 0:
            v = - dspUtil.rmsToDb(np.abs(x))
        elif x == 0:
            v = 0
        else :
            v = dspUtil.rmsToDb(np.abs(x))
        dbs.append(v)
    return dbs
  

#Draft

snd = parselmouth.Sound("ses01.wav")
num_frames = snd.get_number_of_frames()


# Get Jittle
def calculateJitter(data):
    n = len(data)
    sum1 = 0
    sum2 = 0
    for i in range(n):
        if i > 0:
            sum1 += abs(data[i-1] - data[i])
        sum2 += data[i]
    sum1 /= float(n - 1)
    sum2 /= float(n)
    return 100 * sum1 / sum2

def calculateJitterFrequency(data):
    n = len(data)
    dataF = np.zeros(n)
    for i in range(n):
        # convert from periods to F0 per cycle
        dataF[i] = 1.0 / data[i]
    sum1 = 0
    sum2 = 0
    for i in range(n):
        if i > 0:
            sum1 += abs(dataF[i] - dataF[i-1])
        sum2 += dataF[i]
    sum1 /= float(n - 1)
    sum2 /= float(n)
    return  sum1 / sum2

times = [snd.get_time_from_frame_number(i) for i in range (1, num_frames)]
print("num_time:,", len(times))
Jittle = calculateJitterFrequency(times)
print("This is jittle:", Jittle)


#Get Shimmer
def calculateShimmer(data):
    n = len(data)
    sum1 = 0
    sum2 = 0
    for i in range(n):
        if i > 0:
            sum1 += abs(data[i-1] - data[i])
        sum2 += data[i]
    sum1 /= float(n - 1)
    sum2 /= float(n)
    return sum1 / sum2

P = snd.to_intensity(minimum_pitch = 100)
I = [P.get_value(t) for t in times]
I = np.array(I)
I_b = ~np.isnan(I)
I = I[I_b]
Shimmer = calculateShimmer(I)
print("Shimmer: ", Shimmer)


#Get unvoice percentage
Pitch = snd.to_pitch()
f0s = Pitch.selected_array['frequency']
print(len(f0s))
v = Pitch.get_number_of_frames()
print(v)
times = [snd.get_time_from_frame_number(i+1)   for i in range (1, v)]
print(f0s[0:10])
print(times[0:10])
    




# I = snd.to_intensity()
# F = snd.to_formant_burg()
# tt = snd.get_time_step()
#print("time steps: ", snd.n_frames)
#print(F)
# fs = F.get_bandwidth_at_time(1,1)
# fs1 = F.get_value_at_time(2,1)
# print(fs, fs1)


# I1 = snd.get_intensity()
#E1 = snd.ge
#print(I1, E1)
#print("Intensity: \n", len(I))
#print(I.as_array())



# Get f0
f0 = dspUtil.calculateF0once(amp2Db(voiced.get_array_of_samples()), rate)    
#print(f0)

    
#Get mean spectral energy
    
    

#Get vocal intensity




#Get f1, f2, f3 and their bandwidth



#Get Jitter: Irregularity of f0 



#Get shimmer: Irregularity in intensity


#Get percentage of unvoiced period

#Get percentage of breaks

#Get maxDurPause and the average: the longest and average of  pausing period,













```
